```python
df1 = df[['English (en)','Sadness']]
df1 = df1[(df1['Sadness']==1)]

df1
```

```python
Joy = df1.sample(n = 500)
Joy = Joy["English (en)"]
Joy = Joy.tolist()
with open("/content/drive/MyDrive/Colab_Notebooks/lyrics_generator/bag of words/sadness.txt", 'w') as file:
  for index, el in enumerate(Joy):
    if index != len(Joy) - 1:
      file.write(el+"\n")
    else:
      file.write(el)
  file.close()
```
