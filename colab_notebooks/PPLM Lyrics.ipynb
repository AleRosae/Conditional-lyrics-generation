Let's run some examples. The seed number is essential to have different output while using the same input (both for unperturbed and perturbed texts). 

```python
from contextlib import redirect_stdout

run = "fantasy"

with open('/content/'+run+'_run.txt', 'w') as f:
    with redirect_stdout(f):
      run_pplm_example(
          pretrained_model="/content/drive/MyDrive/Colab_Notebooks/lyrics_generator/lyrics/General/training/ultimaprova",
          cond_text="[verse 1]",
          num_samples=3,
          bag_of_words='/content/PPLM/paper_code/wordlists/'+run+'.txt',
          length=100,
          stepsize=0.04,
          sample=True,
          num_iterations=7,
          window_length=5,
          gamma=1.5,
          gm_scale=0.85,
          kl_scale=0.02,
          verbosity='regular',
          seed = 988,
          temperature = 1.3
      )

with open('/content/PPLM/paper_code/wordlists/'+run+'.txt', 'r') as file:
  text = file.readlines()
  with open('/content/'+run+'_run.txt', 'r') as f:
    lyrics = f.read()
    tmp = lyrics.split("================================================================================")
    lyrics = tmp[1]
    #lyrics =  re.sub('\n{2,}',  '\n', lyrics) #clear more than 1 linebreak
    for line in text:
      if line.strip().lower() in lyrics.lower():
        lyrics = re.sub(line.strip(), line.strip().upper(), lyrics)
    with open ("/content/drive/MyDrive/Colab_Notebooks/lyrics_generator/PPLM/"+run+"_run_verse_02.txt", "w") as fe:
      fe.write(lyrics)
```

```python
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    return score
```

```python
from contextlib import redirect_stdout

with open('positive_run.txt', 'w') as f:
    with redirect_stdout(f):
      run_pplm_example(
          pretrained_model="/content/drive/MyDrive/Colab_Notebooks/lyrics_generator/lyrics/General/model",
          cond_text="[Verse 1]>",
          num_samples=3,
          discrim='sentiment',
          class_label='very_positive',
          length=100,
          stepsize=0.05,
          sample=True,
          num_iterations=10,
          gamma=1,
          gm_scale=0.9,
          kl_scale=0.02,
          verbosity='regular'
      )

with open('positive_run.txt', 'r') as f:
  lyrics = f.read()
  tmp = lyrics.split("================================================================================")
  lyrics = tmp[1]
  lyrics =  re.sub('\n{2,}',  '\n', lyrics) #clear more than 1 linebreak
  score = sentiment_analyzer_scores(lyrics)
  with open ("/content/drive/MyDrive/Colab_Notebooks/lyrics_generator/PPLM/very_positive_verse.txt", "w") as f:
    f.write(lyrics)
    f.write(str(score))
```
